# Get outliers

```python
sns.boxplot(dataset_diabetes.Age)
```

```python
def check_outliers_iqr(customers,col):
    q1 = customers[col].quantile(0.25)
    q3 = customers[col].quantile(0.75)
    iqr = q3 - q1
    multiplier = 1.5

    outliers_iqr = customers[
        (customers[col] < (q1 - (multiplier * iqr)))
        | (customers[col] > (q3 + (multiplier * iqr)))
    ]
    return outliers_iqr[[col]]
```

```python
check_outliers_iqr(dataset_diabetes,"Age")
```

```python
## if needed replace outliers with fences

def replace_outliers_with_fences(df, col_name):
    q1 = df[col_name].quantile(0.25)
    q3 = df[col_name].quantile(0.75)
    iqr = q3 - q1
    fence_low = q1 - 1.5 * iqr
    fence_high = q3 + 1.5 * iqr

    df[col_name] = np.where(df[col_name] < fence_low, fence_low, df[col_name])
    df[col_name] = np.where(df[col_name] > fence_high, fence_high, df[col_name])
    return df
```

```python
dataset_diabetes=replace_outliers_with_fences(dataset_diabetes,"Age")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder , StandardScaler , LabelEncoder , RobustScaler
from sklearn.impute import SimpleImputer , KNNImputer
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB , GaussianNB
from sklearn.model_selection import train_test_split , cross_validate
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor,XGBClassifier, plot_importance
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

```

```python
# Instead of:
# Num_Columns = x.select_dtypes(include="number")
# Cat_Columns = x.select_dtypes(include="object_")

# Use the original DataFrame (dataset_diabetes) to select columns:
Num_Columns = dataset_diabetes.select_dtypes(include="number").drop(columns=['Outcome'])
Cat_Columns = dataset_diabetes.select_dtypes(include="object_")
```

```python
Num_Columns.columns
```

```python
Cat_Columns.columns
```

```python
Num_Steps = list()
Num_Steps.append(("Num_Imputer" , KNNImputer()))
Num_Steps.append(("Scaler" , StandardScaler()))
Num_Pipeline = Pipeline(steps=Num_Steps)
```

```python
Cat_Steps = list()
Cat_Steps.append(("Cat_Imputer" , SimpleImputer(strategy='most_frequent')))
Cat_Steps.append(("Cat_Encoder" , OneHotEncoder(sparse_output=False , drop='first'))) # Replace 'sparse' with 'sparse_output'
Cat_Pipeline = Pipeline(steps= Cat_Steps)
```

```python
Transformer = ColumnTransformer(transformers=[('Num' , Num_Pipeline , Num_Columns.columns) ,
                                              ('Cat' , Cat_Pipeline , Cat_Columns.columns) ] , remainder='passthrough')
```

```python
steps = list()
steps.append(("Preprocessing" , Transformer))
steps.append(("Model" , LogisticRegression()))
pipeline = Pipeline(steps=steps)
```

```python
results = cross_validate(pipeline , x , y , scoring="accuracy"  ,cv = 20 , return_train_score=True)
```

# cluster with Kmean

```python
from sklearn.cluster import KMeans
X=dataset_diabetes.drop('Outcome',axis=1)

```

```python
sns.pairplot(X, vars=X.columns)
plt.show()
```

```python
kmeans=KMeans(n_clusters=2,random_state=42)
kmeans.fit(X)
```

```python
cluster_labels  = kmeans.predict(X)
```

```python
dataset_diabetes['Cluster'] = cluster_labels
```

# Visualize the Clusters

```python

plt.scatter(dataset_diabetes.iloc[:, 0], dataset_diabetes.iloc[:, 1], c=dataset_diabetes['Cluster'], cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')
plt.show()
```

```python
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(X)
X['Cluster'] = clusters
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# visualiztion in 3D
